```python
# ì›¹ì—ì„œ ë°ì´í„°ë¥¼ ìˆ˜ì§‘í•´ 7ì¢…ì˜ ì‹ ë°œì— ê´€í•œ ì •ë³´ë¥¼ ëª¨ì•˜ìŠµë‹ˆë‹¤.
# ì°¨ê·¼ì°¨ê·¼ ë°ì´í„° ì „ì²˜ë¦¬ë¥¼ ì‹œë„í•´ë´…ì‹œë‹¤.

name = ['Hoka Clifton 8', 'Brooks Adrenaline GTS 22',
        'Ryka Women\'s Devotion Plus 4 Walking Shoe',
        'Nike Women\'s Pegasus 39', 'Asics Gel-Excite 9',
        'Hoka Women\'s Bondi 8', 'Asics Women\'s GT-2000 10 Running Shoes']
material = ['Mesh', 'Recycled plastic bottle', 'mesh', 'MESH', 'Mesh', 'Mesh', 'Knit']
weight = ['7.2 ounces', '10.2 ounces', '7.4 ounces',
          '9.6 ounces', '7.7 ounces', '8.9 ounces', '8.4 ounces']
price = ['$140', '$132', '$92', '$130', '$75', '$165', '$130']
```

```python
'''
### ì¡°ê±´

1. `material` ë¦¬ìŠ¤íŠ¸ì˜ ì›ì†Œë“¤ì„ ì†Œë¬¸ìë¡œ ë³€ê²½í•˜ê¸°
    - ì†Œì¬ëŠ” í¬ê²Œ 3ì¢…ë¥˜ì˜ ë°ì´í„°ë¡œ ì´ë£¨ì–´ì ¸ ìˆìŠµë‹ˆë‹¤. (ë²”ì£¼í˜• ë³€ìˆ˜)
    í•˜ì§€ë§Œ ë©”ì‰¬ ì†Œì¬ì˜ ê²½ìš° `mesh` , `MESH` , `Mesh` ë¡œ ë‹¤ì–‘í•˜ê²Œ í‘œí˜„ë˜ê³  ìˆìŠµë‹ˆë‹¤.
    3ê°€ì§€ì˜ í‘œê¸°ë²•ì„ í†µì¼ì‹œì¼œ ì£¼ì–´ì•¼ í•˜ê² ë„¤ìš”!
    
2. `weight` ë¦¬ìŠ¤íŠ¸ì˜ â€˜ouncesâ€™ ì œê±°í•˜ê³  ìˆ«ìë§Œ ë‚¨ê¸°ê¸°
    - ë¬´ê²Œ ì •ë³´ëŠ” ìˆ«ìë¡œ ì´ë£¨ì–´ì ¸ ìˆìŠµë‹ˆë‹¤. (ì—°ì†í˜• ë³€ìˆ˜)
    í•˜ì§€ë§Œ ìˆ˜ì§‘í•œ ë°ì´í„°ì—ì„œëŠ” â€˜ouncesâ€™ë¼ëŠ” ë‹¨ìœ„ê°€ ë¶™ì–´ ìˆ«ì ìë£Œí˜•ìœ¼ë¡œ ì·¨ê¸‰í•  ìˆ˜ ì—†ëŠ” ìƒíƒœì…ë‹ˆë‹¤.
    â€˜ouncesâ€™ë¥¼ ì œê±°í•˜ê³  ìˆ«ìë§Œ ë‚¨ê²¨ì„œ ë‹¤ë£¨ê¸° ì‰½ê²Œ ë³€í˜•ì‹œì¼œ ë´…ì‹œë‹¤.
    
3.  `price` ë¦¬ìŠ¤íŠ¸ì˜ â€˜$â€™ ì œê±°í•˜ê³  ìˆ«ìë§Œ ë‚¨ê¸°ê¸°
    - ê°€ê²© ì •ë³´ë„ ë§ˆì°¬ê°€ì§€ë¡œ ì—°ì†í˜• ë³€ìˆ˜ì…ë‹ˆë‹¤. â€˜$â€™ë¥¼ ì œê±°í•˜ê³  ìˆ«ìë§Œ ë‚¨ê²¨ì„œ ë‹¤ë£¨ê¸° ì‰½ê²Œ ë³€í˜•ì‹œì¼œ ë´…ì‹œë‹¤.
    
4. `brand` ë¦¬ìŠ¤íŠ¸ ë§Œë“¤ê¸°
    - ìˆ˜ì§‘í•œ ë°ì´í„°ë¡œë¶€í„° ë‹¤ì–‘í•œ ë°ì´í„°ì˜ í‘œí˜„ì„ ì´ëŒì–´ë‚´ëŠ” ê²ƒë„ ë•Œë¡œëŠ” ì¤‘ìš”í•©ë‹ˆë‹¤.
    í˜„ì¬ ë°ì´í„°ì—ì„œëŠ” ì´ë¦„ ì •ë³´ë¡œë¶€í„° ì‹ ë°œ ë¸Œëœë“œë¥¼ ì¶”ì¸¡í•  ìˆ˜ ìˆì„ ê²ƒ ê°™ìŠµë‹ˆë‹¤.
    ì´ë¦„ì˜ ì²« ë²ˆì§¸ ë‹¨ì–´ë§Œ ê°€ì§€ê³  `brand` ë¼ëŠ” ìƒˆë¡œìš´ ì •ë³´ë¥¼ ë§Œë“¤ì–´ ë´…ì‹œë‹¤.
'''
```

```python
# ë””í´íŠ¸ ë°ì´í„°
name = ['Hoka Clifton 8', 'Brooks Adrenaline GTS 22',
        "Ryka Women's Devotion Plus 4 Walking Shoe",
        "Nike Women's Pegasus 39", 'Asics Gel-Excite 9',
        "Hoka Women's Bondi 8", "Asics Women's GT-2000 10 Running Shoes"]
material = ['Mesh', 'Recycled plastic bottle', 'mesh', 'MESH', 'Mesh', 'Mesh', 'Knit']
weight = ['7.2 ounces', '10.2 ounces', '7.4 ounces',
          '9.6 ounces', '7.7 ounces', '8.9 ounces', '8.4 ounces']
price = ['$140', '$132', '$92', '$130', '$75', '$165', '$130']

# 1. `material` ë¦¬ìŠ¤íŠ¸ì˜ ì›ì†Œë“¤ì„ ì†Œë¬¸ìë¡œ ë³€ê²½í•˜ê¸°
# .lower()ë¥¼ ì‚¬ìš©í•˜ì—¬ ëŒ€ë¬¸ìë¥¼ ëª¨ë‘ ì†Œë¬¸ìë¡œ ë°”ê¿”ì¤Œ
material_filter = [m.lower() for m in material]

# 2. `weight` ë¦¬ìŠ¤íŠ¸ì˜ â€˜ouncesâ€™ ì œê±°í•˜ê³  ìˆ«ìë§Œ ë‚¨ê¸°ê¸°
# float()ë¥¼ ì‚¬ìš©í•˜ì—¬ ë¬¸ìì—´ì„ ì‹¤ìˆ˜í˜•ìœ¼ë¡œ í˜•ë³€í™˜
# .replace()ë¥¼ ì‚¬ìš©í•˜ì—¬ ê´„í˜¸ì™€ ouncesë¥¼ ì œê±°í•¨
weight_filter = [float(w.replace(' ounces', '')) for w in weight]

# 3. `price` ë¦¬ìŠ¤íŠ¸ì˜ â€˜$â€™ ì œê±°í•˜ê³  ìˆ«ìë§Œ ë‚¨ê¸°ê¸°
# int()ë¥¼ ì‚¬ìš©í•˜ì—¬ ë¬¸ìì—´ì„ ì •ìˆ˜í˜•ìœ¼ë¡œ í˜•ë³€í™˜
price_filter = [int(p.replace('$', '')) for p in price]

# 4. `brand` ë¦¬ìŠ¤íŠ¸ ë§Œë“¤ê¸°
# .split()ì„ ì‚¬ìš©í•´ ê³µë°±ì„ ê¸°ì¤€ìœ¼ë¡œ ë¦¬ìŠ¤íŠ¸ ìƒì„± í›„ ì²« ë²ˆì§¸ ì´ë¦„ë§Œ ì¶”ì¶œ
brand = [n.split()[0] for n in name]

# ê²°ê³¼ ì¶œë ¥
print("Material:", material_filter)
print("Weight:", weight_filter)
print("Price:", price_filter)
print("Brand:", brand)
```

```python
'''
### step-up

- ë°ì´í„°ê°€ ì—¬ëŸ¬ ë¦¬ìŠ¤íŠ¸ë¡œ í©ì–´ì ¸ ìˆìœ¼ë‹ˆ í•œ ë²ˆì— ë‹¤ë£¨ê¸° ë¶ˆí¸í•©ë‹ˆë‹¤.
    > ë°ì´í„°ë¥¼ ë”•ì…”ë„ˆë¦¬ë¡œ ë°”ê¾¸ì–´ì„œ ë‹¤ë¤„ë´…ì‹œë‹¤. (í˜¹ì€ íŒë‹¤ìŠ¤ ë°ì´í„°í”„ë ˆì„ìœ¼ë¡œ ë‹¤ë¤„ë´…ì‹œë‹¤.)
- ì‹ ë°œì„ ê°€ë²¼ìš´ ìˆœì„œëŒ€ë¡œ í™•ì¸í•˜ê³  ì‹¶ìŠµë‹ˆë‹¤.
    > `weight` ì— ëŒ€í•´ ì˜¤ë¦„ì°¨ìˆœìœ¼ë¡œ ì •ë ¬í•˜ê³  ê°€ë²¼ìš´ ê²ƒë¶€í„° ì´ë¦„ì„ ì¶œë ¥í•´ë´…ì‹œë‹¤.
- ê°€ê²© ìˆœìœ¼ë¡œë„ í™•ì¸í•˜ê³  ì‹¶ìŠµë‹ˆë‹¤.
    > `price` ì— ëŒ€í•´ ì˜¤ë¦„ì°¨ìˆœìœ¼ë¡œ ì •ë ¬í•˜ê³  ì €ë ´í•œ ê²ƒë¶€í„° ì´ë¦„ì„ ì¶œë ¥í•´ë´…ì‹œë‹¤.
'''
```

```python
import pandas as pd  # íŒë‹¤ìŠ¤ ë¶ˆëŸ¬ì˜¤ê¸°

# ì •ì œëœ ë¦¬ìŠ¤íŠ¸ë“¤
name = ['Hoka Clifton 8', 'Brooks Adrenaline GTS 22',
        "Ryka Women's Devotion Plus 4 Walking Shoe",
        "Nike Women's Pegasus 39", 'Asics Gel-Excite 9',
        "Hoka Women's Bondi 8", "Asics Women's GT-2000 10 Running Shoes"]
material = ['Mesh', 'Recycled plastic bottle', 'mesh', 'MESH', 'Mesh', 'Mesh', 'Knit']
material_cleaned = [m.lower() for m in material]

weight = ['7.2 ounces', '10.2 ounces', '7.4 ounces', '9.6 ounces', '7.7 ounces', '8.9 ounces', '8.4 ounces']
weight_cleaned = [float(w.replace(' ounces', '')) for w in weight]

price = ['$140', '$132', '$92', '$130', '$75', '$165', '$130']
price_cleaned = [int(p.replace('$', '')) for p in price]

brand = [n.split()[0] for n in name]

# ë”•ì…”ë„ˆë¦¬ë¥¼ ë§Œë“¤ì–´ DataFrameìœ¼ë¡œ ë³€í™˜
data = pd.DataFrame({
    'name': name,
    'brand': brand,
    'material': material_cleaned,
    'weight': weight_cleaned,
    'price': price_cleaned
})

# ë°ì´í„°í”„ë ˆì„ í™•ì¸
print(data)

# weight ê¸°ì¤€ìœ¼ë¡œ ì˜¤ë¦„ì°¨ìˆœ ì •ë ¬
sorted_by_weight = data.sort_values(by='weight', ascending=True)

# ê°€ë²¼ìš´ ìˆœì„œëŒ€ë¡œ ì´ë¦„ë§Œ ì¶œë ¥
print("ğŸƒ ê°€ë²¼ìš´ ìˆœì„œë¡œ ì‹ ë°œ ì´ë¦„:")
print(sorted_by_weight['name'].to_list())

# price ê¸°ì¤€ìœ¼ë¡œ ì˜¤ë¦„ì°¨ìˆœ ì •ë ¬
sorted_by_price = data.sort_values(by='price', ascending=True)

# ì €ë ´í•œ ìˆœì„œëŒ€ë¡œ ì´ë¦„ë§Œ ì¶œë ¥
print("ğŸ’° ì €ë ´í•œ ìˆœì„œë¡œ ì‹ ë°œ ì´ë¦„:")
print(sorted_by_price['name'].to_list())
```
