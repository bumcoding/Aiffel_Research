```python
'''
### 외계 행성 B-17의 생물 특징 파악하기

- 여러분은 아주 비밀스러운 의뢰를 받았습니다.
- 외계 행성 B-17을 탐험했던 이들로부터 행성에 살고 있는 생물들을 분석해달라는 의뢰였습니다.
- 아주 기초적인 분석부터 시작해봅시다!

여러분은 의뢰인으로부터 두 개의 파일을 전달 받았습니다.

- creatures.csv : 행성 B-17에 사는 생물들의 기본 정보입니다.
    - id : 생물의 고유 ID
    - species : 종 이름
    - weight : 몸무게 (kg)
    - height : 키 (cm)
    - age : 나이 (년)
    
- environment.csv : 생물들이 사는 환경 데이터입니다.
    - id : 생물이 서식하는 환경의 ID
    - temperature : 평균 기온 (°C)
    - humidity : 평균 습도 (%)
    - habitat : 서식지의 이름 (예: 숲, 사막, 바다)

### 조건

1. 두 csv 파일을 `pandas`를 이용해 `DataFrame` 으로 불러오세요
2. 각 종(`species`) 별로 생물들의 평균 몸무게와 키를 계산하고 결과를 출력하세요.
3. 각 종(`species` )별로 생물 나이의 중앙값을 계산하고 결과를 출력하세요.
4. 나이가 많은 순서대로 데이터를 정렬해보고, 가장 나이가 많은 생물의 정보(`species`, `weight` , `height`, `age` )를 출력하세요
5. 두 데이터(creature.csv, enviorment.csv)를 `id` 기준으로 합쳐주세요.
6. 생물이 서식하는 환경(`habitat`) 별로 평균 몸무게와 키를 계산하고 결과를 출력하세요.
7. 생물이 서식하는 환경의 평균 습도(`humidity`)와 생물의 체중(`weight`) 간의 상관계수를 계산해 보세요.
8. 각 생물에 대해 체중 대비 키 비율(`height / weight`)을 계산하고 값이 가장 높은 생물의 정보를 출력하세요.
9. 평균 기온(`temperature`)이 0°C ~ 14°C, 15°C ~ 29°C, 30°C 이상인 서식지로 그룹을 나누고, 각 그룹별로 생물의 평균 몸무게와 키를 계산하세요.
'''
```

```python
import pandas as pd

# 1. 두 csv 파일을 불러오기
creatures = pd.read_csv("/Users/bumchanpark/Desktop/Aiffel_Research/tutorial/tutorial6/creatures.csv")
environment = pd.read_csv("/Users/bumchanpark/Desktop/Aiffel_Research/tutorial/tutorial6/environment.csv")

# 2. 'species' 별로 생물들의 평균 몸무게와 키
# .groupby()를 통해 종을 기준으로 몸무게와 키의 평균값을 구함
species_avg = creatures.groupby("species")[["weight", "height"]].mean()
print(species_avg, "\n")

# 3. 'species' 별로 생물 나이의 중앙값
# .median()은 중앙값을 계산하는 함수
age = creatures.groupby("species")["age"].median()
print(age, "\n")

# 4. 나이 많은 순으로 정렬하고 가장 나이 많은 생물 정보
# .sort_values()은 오름차순 정렬 함수로 내림차순을 하고 싶으면 ascending = False를 포함함
# .iloc[0]으로 가장 먼저 오는 값(가장 큰 수)를 추출함
old = creatures.sort_values(by = "age", ascending = False).iloc[0]
print(old[["species", "weight", "height", "age"]], "\n")

# 5. 두 데이터를 'id' 기준으로 합치기
# pd.merge()를 통해 두 데이터프레임을 id를 기준으로 합친다 (on = "합치는 기준")
combine = pd.merge(creatures, environment, on = "id")
print(combine.head(), "\n")

# 6. 'habitat' 별로 평균 몸무게와 키
habitat_avg = combine.groupby("habitat")[["weight", "height"]].mean()
print(habitat_avg, "\n")

# 7. 습도와 체중 간의 상관계수
# .corr()를 통해 목록 간의 상관계수를 계산함
correlation = combine["humidity"].corr(combine["weight"])
print(correlation, "\n")

# 8. 키와 몸무게 비율이 가장 높은 생물의 정보
# 데이터프레임에 "ratio"행 추가 (키와 몸무게의 비율 정보)
combine["ratio"] = combine["height"] / combine["weight"]
high = combine.sort_values(by = "ratio", ascending = False).iloc[0]
print(high[["species", "weight", "height", "age", "ratio"]], "\n")

# 9. 온도 구간별로 그룹 나누고 평균 몸무게와 키
# 기온에 따른 그룹을 나누는 함수 정의
def temperature_group(temp):
    # 주어진 조건을 만족하는 if문
    if temp < 15:
        return "0-14°C"
    elif temp < 30:
        return "15-29°C"
    else:
        return "30°C~"
    
# 각 생물에 해당하는 값을 .apply()로 함수를 불러옴
combine["temp_group"] = combine["temperature"].apply(temperature_group)
temp_avg = combine.groupby("temp_group")[["weight", "height"]].mean()
print(temp_avg)
```

```python
'''
### step-up

- 별 평균 몸무게와 키를 막대 그래프로 시각화해보세요.
- 숫자로 데이터를 파악하는 것도 좋지만, 시각화를 통해 데이터를 분석하면 더 직관적인 해석이 가능합니다.
- 서식지 별 평균 몸무게와 키를 막대 그래프로 시각화해보세요.
- 생물들의 나이대 분포를 히스토그램으로 시각화해보세요.
- 생물들의 몸무게와 키를 산점도(scatter plot)으로 시각화해보세요. 종별로 색을 달리해서 시각화해보세요.
'''
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ✅ 한글 폰트 설정
# macOS 사용자는 AppleGothic, Windows는 Malgun Gothic 사용
plt.rcParams['font.family'] = 'AppleGothic'     # macOS
# plt.rcParams['font.family'] = 'Malgun Gothic' # Windows일 경우 주석 해제
plt.rcParams['axes.unicode_minus'] = False      # 마이너스 기호 깨짐 방지

# 1. CSV 파일 불러오기
creatures = pd.read_csv("/Users/bumchanpark/Desktop/Aiffel_Research/tutorial/tutorial6/creatures.csv")
environment = pd.read_csv("/Users/bumchanpark/Desktop/Aiffel_Research/tutorial/tutorial6/environment.csv")
merged = pd.merge(creatures, environment, on="id")

# 2. 종(species)별 평균 몸무게와 키 시각화
species_group = creatures.groupby("species")[["weight", "height"]].mean().reset_index()

plt.figure(figsize=(10, 5))
species_group.plot(kind="bar", x="species", y=["weight", "height"], colormap="Paired")
plt.title("종별 평균 몸무게와 키")
plt.ylabel("값")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 3. 서식지(habitat)별 평균 몸무게와 키 시각화
habitat_group = merged.groupby("habitat")[["weight", "height"]].mean().reset_index()

plt.figure(figsize=(10, 5))
habitat_group.plot(kind="bar", x="habitat", y=["weight", "height"], colormap="Set2")
plt.title("서식지별 평균 몸무게와 키")
plt.ylabel("값")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 4. 생물 나이 분포 히스토그램
plt.figure(figsize=(8, 5))
sns.histplot(creatures["age"], bins=15, kde=True, color="skyblue")
plt.title("생물 나이 분포")
plt.xlabel("나이 (년)")
plt.ylabel("개체 수")
plt.tight_layout()
plt.show()

# 5. 몸무게 vs 키 산점도 (종별 색상)
plt.figure(figsize=(8, 6))
sns.scatterplot(data=creatures, x="weight", y="height", hue="species", palette="tab10", s=60)
plt.title("몸무게와 키 관계 (종별 색상)")
plt.xlabel("몸무게 (kg)")
plt.ylabel("키 (cm)")
plt.tight_layout()
plt.show()
```
